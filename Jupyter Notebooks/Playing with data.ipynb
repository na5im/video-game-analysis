```python
import pandas as pd
df = pd.read_csv("games_with_units_shipped.csv")
df = df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
df
```

```python
ndf = df.loc[df["units_shipped"].isnull() != True]
ndf = ndf.sort_values(['franchise', 'name'], ascending=[True, False])
ndf
```

```python
ndf.to_csv("games_with_missing.csv")
```

```python
newdf = pd.read_csv("games_with_units_shipped2.csv")
newdf = newdf.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
nndf = newdf.loc[newdf["units_shipped"].isnull() != True]
nndf
```

```python
combined_df = pd.concat([ndf, nndf], ignore_index=True)
combined_df
```

```python
import re
new_df = combined_df
names = []
new_df['cleaned_names'] = new_df['name'].apply(lambda x: re.sub(r'\W+', ' ', x))

for x in new_df['cleaned_names']:
    names.append(x)
names
```

```python
na_df = newdf.loc[newdf["units_shipped"].isnull() == True]
na_df.to_csv('no_units.csv')
```

```python
merged_df = pd.concat([combined_df, na_df], ignore_index=True)
merged_df = merged_df.drop(columns=['cleaned_name', 'cleaned_names'])
```

```python
cleaned_df = pd.read_csv("cleaned_games.csv")
cleaned_df.drop('units_shipped', axis=1, inplace=True)
cleaned_df
```

```python
columns_to_check = ['aggregated_rating','total_rating','follows','hypes']

updated_df = pd.concat([combined_df, cleaned_df], ignore_index=True).drop(columns=['cleaned_names','cleaned_name', 'Column'])
filtered_df = updated_df[updated_df[columns_to_check].isnull().all(axis=1) == False]
filtered_df
```

```python
final = pd.read_csv('final_clean.csv')
final
```

```python
final['count_non_null'] = final[columns_to_check].notnull().sum(axis=1)
final_df = final[final['count_non_null'] >= 2]
final_df = final_df.drop(columns='Column')
final_df
```

```python
still_no_units = final_df.loc[final_df['units_shipped'].isnull() == True]
still_no_units
```
